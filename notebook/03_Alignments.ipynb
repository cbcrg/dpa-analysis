## Alignments
The alignments generated below are used both as guide trees in the progressive (standard)
alignment proceedure as well as trees used in the clustering step of the regressive alignment 
proceedure.

There is an additional set of command line for the default alignments where no guide tree 

The command lines can be found below and in the `templates/trees` directory of the repository.

The generated trees can be found in the `data/trees` directory of this repository.
